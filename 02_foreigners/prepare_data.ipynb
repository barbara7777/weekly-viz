We need to do some cleaning

```python
country_map = {"Država državljanstva - SKUPAJ": "TOGETHER",
               "Države EU": "EU", 
               "... druge države članice EU": "EU",
               "... Hrvaška": "Croatia",
               "Hrvaška": "Croatia",
               "Bosna in Hercegovina": "Bosnia",
               "Srbija": "Serbia & Montenegro",
               "Srbija in Črna gora": "Serbia & Montenegro",
               "Črna gora": "Serbia & Montenegro",
               "Kosovo": "Kosovo",
               "Druge evropske države": "Other European Countries",
               "Neevropske države": "Noneuropean countries",
               "Neznano": "Unknown"
                }           
```

```python
def trnasform_day_to_numeric(val):
    return int(val) if val.isdigit() else np.nan

data_df["LETO"] = data_df["LETO"].astype(int)
data_df["DATA"] = data_df["DATA"].apply(trnasform_day_to_numeric)
data_df["NAMEN PRISELITVE"] = pd.Categorical(data_df["NAMEN PRISELITVE"])
data_df["DRŽAVA DRŽAVLJANSTVA"] = data_df["DRŽAVA DRŽAVLJANSTVA"].apply(lambda val: country_map[val])
```

```python
data_df = data_df.rename(columns={"LETO": "year", "DATA": "immigrations", "NAMEN PRISELITVE": "purpose of immigration", "DRŽAVA DRŽAVLJANSTVA": "citizenship"})
```

```python
data = data_df.dropna()
```

```python
# remove rows with grouped counts
data = data[~data["purpose of immigration"].str.contains("SKUPAJ")]
```
